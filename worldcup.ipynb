BATTING Analysis

```python
df.columns
```

```python
Total_runs = df['runs_off_bat'].sum()  + df['extras'].sum() + df['wides'].sum() + df['noballs'].sum() + df['byes'].sum() + df['legbyes'].sum() + df['penalty'].sum()
print('Total runs in wc' , Total_runs)#total runs in wc
```

```python
extras = df['extras'].sum() + df['wides'].sum() + df['noballs'].sum() + df['byes'].sum() + df['legbyes'].sum() + df['penalty'].sum()
print('Total  exttras runs in wc' ,extras) #extras 
```

```python
pd.DataFrame(df.groupby('striker')['runs_off_bat'].sum().sort_values(ascending=False).head(5))#highest run scorer
```

```python
pd.DataFrame(df.groupby('batting_team')['runs_off_bat'].count().sort_values(ascending=False))# highwst runs by country
```

```python
pd.DataFrame(df.groupby(['match_id', 'innings'])['runs_off_bat'].mean().round(2)).head(10)#round used for 2 digit or 3 digit value #2 average run per innings
```

```python
df.groupby(['match_id','innings'])['runs_off_bat'].sum().mean() # average score per inning
```

```python
#highest individual inning score
highest_individual_score= df.groupby(['match_id','striker'])['runs_off_bat'].sum().reset_index()
max_score_in_each = highest_individual_score.groupby('match_id')['runs_off_bat'].max().sort_values(ascending = False)
dff = pd.DataFrame(max_score_in_each)
dff.head()
```

```python
pd.DataFrame(df.groupby(['match_id'])['runs_off_bat'].sum().sort_values(ascending=False))
```

Toss Analysis

```python
match = pd.read_csv('matches.csv')
```

```python
match.head()
```

```python
match.columns
```

```python
#percentage of team won match who won toss
count = 0
for index, row in match.iterrows():
    if row['toss_winner'] == row['winner']:
        count += 1
per = round((count / match['match_number'].count()) * 100 , 2)
print('The percentage of matches won by the team that won the toss ',per)
```

```python
match.groupby('toss_decision')['match_number'].count()#toss desisions
```

```python
match.groupby(['venue','toss_decision']).size().reset_index(name='count')#toss desision acroes venues
```

Umpire Analysis

```python
count = pd.concat([match['umpire1'], match['umpire2'], match['reserve_umpire']]).nunique()
print('Total number of umpires in wc are' ,count)
```
